```python
#carga de datos
df_prod = pd.read_csv('learnplatform-covid19-impact-on-digital-learning/products_info.csv')
df_district = pd.read_csv('learnplatform-covid19-impact-on-digital-learning/districts_info.csv')
archivos_data = [os.path.join('learnplatform-covid19-impact-on-digital-learning/engagement_data', f) for f in os.listdir('learnplatform-covid19-impact-on-digital-learning/engagement_data') if f.endswith('.csv')]
df_engagement = pd.concat([pd.read_csv(f).assign(district_id=int(os.path.basename(f)[:-4])) for f in archivos_data])  # Convertir a entero dentro de assign
df_prod.columns = df_prod.columns.str.lower().str.replace(' ', '_')
df_district.columns = df_district.columns.str.lower().str.replace(' ', '_')
df_engagement.columns = df_engagement.columns.str.lower().str.replace(' ', '_')
```

```python
#hacemos merge de los 3 df usando el indice caracteristico
df_merged = pd.merge(df_engagement, df_prod, on='lp_id', how='inner')
df_merged = pd.merge(df_merged, df_district, on='district_id', how='inner')
print('Porcentaje de datos vacios en merge:')
print(df_merged.isnull().sum()/df_merged.shape[0]*100)
```

Notemos que la cantidad de datos vacios no es despreciable. Por lo que se reemplazara con la moda

```python
df_merged.fillna(df_merged.mode().iloc[0], inplace=True)
print('Cantidad de datos vacios en merge:')
print(df_merged.isnull().sum()/df_merged.shape[0]*100)
```

**¿Cuál es el panorama de la conectividad y el compromiso digitales en 2020?**

Analizaremos primero la tendencia en el uso de las herramientas digitales antes, durante y despues del cierre de las escuelas.

```python
df_merged['time'] = pd.to_datetime(df_merged['time'])
engagement_x_tiempo=df_merged.groupby('time')['engagement_index'].mean().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(x='time',y='engagement_index',data=engagement_x_tiempo)
plt.title('Compromiso con el aprendizaje digital a lo largo de tiempo')
plt.show()
```

Es claro que despues del cierro de los colegios aumento el tiempo de uso de las herramientas virtuales. Sin embargo, el uso antes sigue siendo alto.

Veamos ahora la relacion con respecto al distrito.

```python
#ahora separamos por distrito
engagement_by_district = df_merged.groupby('district_id')['engagement_index'].mean().reset_index()
engagement_by_district = pd.merge(engagement_by_district, df_district, on='district_id')
plt.figure(figsize=(12, 6))
sns.boxplot(x='locale', y='engagement_index', data=engagement_by_district)
plt.title('Compromiso con el aprendizaje digital por localidad')
plt.show()
```

En general vemos que en el ambito rural hubo un gran compromiso hacia las herramientas virtuales, ademas, su distribucion es sesgada a los 400 indice de compromiso.
Por otro lado, en la ciudad no hubo mayor cambio, excepto una tendencia a usar estas herramientas, ya que, la mediana esta muy arriba. Ademas, es donde se notan mas datos atipicos.
Por contra, en los suborbios y en los pueblos la tendencia es a no usarlas, aunque si hubo un aumento.

¿Que tipo de productos tuvo mayor impacto en el compromiso durante la pandemia?

```python
engagement_by_product = df_merged.groupby('lp_id')['engagement_index'].mean().reset_index()
engagement_by_product = pd.merge(engagement_by_product, df_prod, on='lp_id')
category_engagement = engagement_by_product.groupby('primary_essential_function')['engagement_index'].mean().reset_index()
#se escoge el top 10, pues, son demasiadas
top_10_categories = category_engagement.sort_values(by='engagement_index', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.heatmap(top_10_categories.pivot_table(index='primary_essential_function', values='engagement_index'), annot=True, cmap='viridis', fmt=".2f")
plt.title('Top 10 de funcionalidad con mayor compromiso')
plt.xlabel('Categoría de funcionalidad')
plt.ylabel('Compromiso promedio')
plt.show()
```

Notemos que tenemos una alta correlacion en LMS, sistema de gestion del aprendizaje, es decir realizar el quiz o la prueba fue lo que mayor compromiso dio. Por ultimo, asistir a clases viruales tuvo una pequeño compromiso con los alumnos. Podemos concluir que el mayor compromiso para y con las herramientas virtuales fue hacer el minimo esfuerzo por aprobar el semestre.

```python
#ahora para productos
plt.figure(figsize=(12, 6))
top_10_products = engagement_by_product.sort_values(by='engagement_index', ascending=False).head(10)
sns.heatmap(top_10_products.pivot_table(index='product_name', values='engagement_index'), annot=True, fmt=".2f")
plt.title('Top 10 Productos con Mayor Compromiso')
plt.xlabel('Nombre del producto')
plt.ylabel('Compromiso promedio')
plt.show()
```

Efectivamente, como se dijo anteriormente. El mayor tiempo que pasaron los alumnos fue en las paginas donde se realizaban quiz, tareas o mandados. El resto de plataformas no se le dio apenas uso.
